# What is the approximate average drug overdose death rate in the United States?

```python
#What is the approximate average drug overdose death rate in the United States?
#Question 1
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data_frame = pd.read_csv('Drug_overdose_death_rates__by_drug_type__sex__age__race__and_Hispanic_origin__United_States (1).csv')
data_frame.drop_duplicates(inplace=True)
data_frame.reset_index(drop=True, inplace=True)
#This step is to import the data in the csv file. If it has been completed before,
#it can be omitted (theoretically the import has been completed according to the steps), 
#so this line of code only gives the notebook the ability to run without following the steps.

average_death_rates = data_frame['ESTIMATE'].mean()
#Calculate average
print(f'The average ESTIMATE value for drug overdose death NUM per 100000 person is: {average_death_rates}')
print("the average percentage of drug overdose deaths in is ",average_death_rates/1000,"%")
print("The approximate average drug overdose death rate in the United States {:.4f}%".format(average_death_rates/1000))
#I use four digits after the decimal point instead of two because the data is too small and will appear as 0
```

Running the code above can calculate the death average from public data to answer question 1 "What is the approximate average drug overdose death rate in the United States?"

# Question NO.2
# Are the number of drug overdose deaths per 100,000 people in the United States related to gender? 
# If so, which gender had more deaths?

```python
#Are the number of drug overdose deaths per 100,000 people in the United States related to gender? 
#If so, which gender had more deaths?
import pandas as pd
import numpy as np
data_frame = pd.read_csv('Drug_overdose_death_rates__by_drug_type__sex__age__race__and_Hispanic_origin__United_States (1).csv')
data_frame.drop_duplicates(inplace=True)
data_frame.reset_index(drop=True, inplace=True)
#This step is to import the data in the csv file. If it has been completed before,
#it can be omitted (theoretically the import has been completed according to the steps), 
#so this line of code only gives the notebook the ability to run without following the steps.
import matplotlib.pyplot as plt
#Question NO.2
male_e_avg = data_frame[data_frame['STUB_LABEL'] == 'Male']['ESTIMATE'].mean()
female_e_avg = data_frame[data_frame['STUB_LABEL'] == 'Female']['ESTIMATE'].mean()
# i am use pie chart here
labels = ['Male', 'Female']
sizes = [male_e_avg, female_e_avg]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=240, colors=['cyan', 'lightcoral'])
plt.title('Overdose deaths per 100,000 people by gender')
plt.show()
print(f"The number of male who die from overdose per 100,000 people is {male_e_avg:.2f}")
print(f"The number of female who die from overdose per 100,000 people is: {female_e_avg:.2f}")
```

After analyzing the data provided by Python and examining the figure presented, it is evident that the percentage of overdose-related deaths among men in the United States is significantly higher than that among women. This discrepancy may be attributed to the fact that men are more likely to use illegal drugs compared to women. 
This data provides us with a more nuanced understanding of drug use and overdose rates in the United States. However, it raises crucial questions about the social and cultural factors contributing to this disparity. # Question 3 Among all types of addictive drugs, which drug is the most harmful in causing overdose death?

```python
# Exclude the "All drug overdose deaths" 
panel_mean = data_frame[data_frame['PANEL'] != 'All drug overdose deaths'].groupby('PANEL')['ESTIMATE'].mean().sort_values(ascending=False)
# find the PANEL with the highest average ESTIMATE
higpanel = panel_mean.idxmax()
higvalue = panel_mean.max()
print(f" which drug is the most harmful in causing overdose death in united state is: {higpanel}")
print(f"The average number of deaths per 100,000 people caused by this drug nationwide is {higvalue:.2f}")
```

The average number of deaths per 100,000 people caused by this type of drug nationwide is opioids.
This data is predictable because opioid drugs were also the most prominent in the previous bar chart,
which illustrates the need for the government to step up the supervision of opioid drugs.

# Question 5：How does the overdose mortality rate change each year?

```python
#How does the overdose mortality rate change each year?
#Question 5
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data_frame=pd.read_csv('Drug_overdose_death_rates__by_drug_type__sex__age__race__and_Hispanic_origin__United_States (1).csv')
data_frame.drop_duplicates(inplace=True)
data_frame.reset_index(drop=True, inplace=True)
#This step is to import the data in the csv file. If it has been completed before,
#it can be omitted (theoretically the import has been completed according to the steps), 
#so this line of code only gives the notebook the ability to run without following the steps.
if data_frame.empty:
  print("The chart generates an error")
else:
  print("Dot plot generated successfully")
# average overdose mortality rate for each year
av_death_rates = data_frame.groupby('YEAR')['ESTIMATE'].mean()
# Sort the years
sorted_years = av_death_rates.sort_values().index
# Create a dot plot
plt.figure(figsize=(10, 6))
plt.plot(av_death_rates.loc[sorted_years], 'bo')
plt.title('Relationship between Year and Average Overdose Mortality Rate')
plt.xlabel('Year')
plt.ylabel('Average drug Overdose Mortality per 100000 in us')
plt.show()

```

# Question 6
# Based on this data set, how do different age groups perform in drug overdose mortality?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data_frame = pd.read_csv('Drug_overdose_death_rates__by_drug_type__sex__age__race__and_Hispanic_origin__United_States (1).csv')
data_frame.drop_duplicates(inplace=True)
data_frame.reset_index(drop=True, inplace=True)
#This step is to import the data in the csv file. If it has been completed before,
#it can be omitted (theoretically the import has been completed according to the steps), 
#so this line of code only gives the notebook the ability to run without following the steps.

age_totals = data_frame[data_frame['AGE'] != 'All ages'].groupby('AGE')['ESTIMATE'].sum()
#pie chart
plt.figure(figsize=(9, 9))
plt.pie(age_totals, labels=age_totals.index, autopct='%1.1f%%', startangle=140)
plt.title('Total death per 100000 in united state by Age Group')
plt.show()

```

In this chart, we can see that the 35-44-year-old group and the 45-55-year-old group have the highest rates of drug overdose deaths in the United States. These groups have more opportunities to be exposed to drugs than older people and also have higher rates of drug overdose than younger people. People are in worse physical condition.
In addition, mental health also affects the probability of overdose. The 35-44-year-old group and the 45-55-year-old group are also high-risk groups for mental illness. It is not difficult to explain such a high proportion.
